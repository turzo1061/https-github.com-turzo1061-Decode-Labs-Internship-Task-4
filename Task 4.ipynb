**Reasoning**:
I need to implement a rank-based ensemble. This involves getting the predicted probabilities from each model, ranking the classes based on these probabilities, summing the ranks for each class across all models, and then making the final prediction based on the lowest summed rank. Finally, I will evaluate the ensemble's performance.



```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Get predicted probabilities for all models
# y_pred_log_reg_proba, y_pred_cat_proba, y_pred_nn_proba are already available

# Combine probabilities for ensemble
# For rank-based ensemble, we will rank probabilities directly

# Function to get ranks for each class based on probabilities
def get_ranks(probabilities):
    # Sort probabilities in descending order and get their ranks
    # Smaller rank means higher probability
    return np.argsort(np.argsort(-probabilities, axis=1), axis=1)

# Get ranks for each model
ranks_log_reg = get_ranks(y_pred_log_reg_proba)
ranks_cat = get_ranks(y_pred_cat_proba)
ranks_nn = get_ranks(y_pred_nn_proba)

# Sum the ranks for each class across all models
summed_ranks = ranks_log_reg + ranks_cat + ranks_nn

# The class with the lowest summed rank is the ensemble's prediction
y_pred_ensemble = np.argmin(summed_ranks, axis=1)

# Calculate accuracy
accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
print(f"Ensemble Model Accuracy: {accuracy_ensemble:.4f}")

# Classification Report
print("\nEnsemble Model Classification Report:")
print(classification_report(y_test, y_pred_ensemble))

# Confusion Matrix
conf_matrix_ensemble = confusion_matrix(y_test, y_pred_ensemble)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_ensemble, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Ensemble Model Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
